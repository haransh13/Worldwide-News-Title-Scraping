```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime

# List of African and Oceanian countries and their respective Google News URLs
countries = {
    # African countries
    'algeria': "https://news.google.com/topics/CAAqJAgKIh5DQkFTRUFvSEwyMHZNR2d6ZVJJRlpXNHRSMElvQUFQAQ?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'angola': "https://news.google.com/topics/CAAqJAgKIh5DQkFTRUFvSEwyMHZNR28wWWhJRlpXNHRSMElvQUFQAQ?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'benin': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREUyTkhZU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'botswana': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREUyTm5ZU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'burkina_faso': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREUyT1RrU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'burundi': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREUyT1hRU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'cape_verde': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREZ1Y1dvU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'cameroon': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREZ1Ykc0U0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'central_african_republic': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREZ1ZVd3U0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'chad': "https://news.google.com/search?q=chad&hl=en-IN&gl=IN&ceid=IN%3Aen",
    'comoros': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREZ1Tm1NU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'congo': "https://news.google.com/search?q=congo&hl=en-IN&gl=IN&ceid=IN%3Aen",
    'congo_dr': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRGc0ZUhBU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'djibouti': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREkzYW1zU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'egypt': "https://news.google.com/search?q=egypt&hl=en-IN&gl=IN&ceid=IN%3Aen",
    'equatorial_guinea': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREpyWTNvU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'eritrea': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNREpyYUhNU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'eswatini': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFoyTXpZU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'ethiopia': "https://news.google.com/topics/CAAqJggKIiBDQkFTRWdvSkwyMHZNREU1Y0dOekVnVmxiaTFIUWlnQVAB?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'gabon': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRE0xTkRnU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'gambia': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR2hrZURnU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'ghana': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRE0xWkdzU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'guinea': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRE0yTnpZU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'guinea_bissau': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRE0yWWw4U0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'ivory_coast': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR1oyTkhZU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'kenya': "https://news.google.com/topics/CAAqJggKIiBDQkFTRWdvSkwyMHZNREU1Y21jMUVnVmxiaTFIUWlnQVAB?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'lesotho': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFJvZG5jU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'liberia': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFJvZW1vU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'libya': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFJuY1hJU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'madagascar': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFJ6YWpNU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'malawi': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFIwY2pFU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'mali': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFIyTURrU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'mauritania': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFIyYW1nU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'mauritius': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFIyY3prU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'morocco': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFIzWjJnU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'mozambique': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFIzYkdnU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'namibia': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFZpYlhFU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'niger': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFZqWXpFU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'nigeria': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFZqWjNZU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'rwanda': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFprWm1jU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'sao_tome_and_principe': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR28yT0RFU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'senegal': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFp6Y21zU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'seychelles': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFp6ZHprU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'sierra_leone': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFp6WHpJU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'somalia': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFowWjNjU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'south_africa': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR2g2YkhvU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'south_sudan': "https://news.google.com/topics/CAAqJggKIiBDQkFTRWdvSkwyMHZNRFZ5ZW01NkVnVmxiaTFIUWlnQVAB?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'sudan': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRFowZHpnU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'tanzania': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRGRrZW1ZU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'togo': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRGRtTlhnU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'tunisia': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRGRtYWw4U0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'uganda': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRGQwY0RJU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'zambia': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRGc0ZG1JU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen",
    'zimbabwe': "https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNRGc0Y1RRU0JXVnVMVWRDS0FBUAE?hl=en-IN&gl=IN&ceid=IN%3Aen"
}

def classify_news(title):
    """
    Classify the news based on the title.

    Args:
        title (str): The title of the news article.

    Returns:
        str: The category of the news article.
    """
    keywords = {
        'finance': ['stock', 'market', 'finance', 'economy', 'investment', 'banking', 'money', 'business', 'trade', 'inflation', 'interest rates'],
        'education': ['school', 'education', 'university', 'college', 'student', 'teacher', 'classroom', 'curriculum', 'scholarship', 'academic', 'degree'],
        'political': ['election', 'government', 'policy', 'political', 'vote', 'legislation', 'senate', 'parliament', 'campaign', 'democracy', 'candidate'],
        'war': ['war', 'Gaza', 'Palestine', 'Palestinian', 'conflict', 'military', 'battle', 'troops', 'army', 'navy', 'air force', 'weapon', 'invasion', 'defense', 'security'],
        'entertainment': ['movie', 'music', 'celebrity', 'entertainment', 'concert', 'film', 'theater', 'festival', 'award', 'show', 'series', 'actor', 'actress'],
        'tech': ['technology', 'tech', 'software', 'gadget', 'innovation', 'AI', 'robotics', 'internet', 'startup', 'app', 'device', 'cybersecurity', 'IT'],
        'god': ['religion', 'faith', 'church', 'god', 'spiritual', 'worship', 'prayer', 'bible', 'holy', 'deity', 'temple', 'pastor', 'imam', 'rabbi'],
        'health': ['health', 'medicine', 'hospital', 'doctor', 'nurse', 'disease', 'virus', 'pandemic', 'treatment', 'vaccine', 'surgery', 'mental health', 'fitness', 'nutrition'],
        'sports': ['sports', 'game', 'match', 'team', 'player', 'tournament', 'championship', 'league', 'score', 'goal', 'coach', 'athlete'],
        'environment': ['environment', 'climate', 'pollution', 'recycle', 'conservation', 'wildlife', 'sustainability', 'natural disaster', 'weather', 'global warming', 'ecology'],
        'crime': ['crime', 'criminal', 'police', 'law enforcement', 'court', 'trial', 'sentence', 'verdict', 'theft', 'robbery', 'assault', 'fraud', 'murder'],
        'science': ['science', 'research', 'experiment', 'discovery', 'space', 'astronomy', 'biology', 'chemistry', 'physics', 'scientist', 'laboratory'],
        'travel': ['travel', 'tourism', 'vacation', 'trip', 'destination', 'hotel', 'flight', 'tourist', 'journey', 'adventure', 'cruise'],
        'fashion': ['fashion', 'style', 'designer', 'clothing', 'runway', 'trend', 'model', 'boutique', 'couture', 'wardrobe'],
    }
    for category, words in keywords.items():
        if any(word in title.lower() for word in words):
            return category
    return 'other'

def scrape_google_news(country, url):
    """
    Scrape news from Google News for a specific country.

    Args:
        country (str): The name of the country.
        url (str): The URL of the Google News page for the country.

    Returns:
        list: A list of dictionaries containing the news data.
    """
    driver = webdriver.Chrome()  # Make sure ChromeDriver is in your PATH or specify the executable_path
    driver.get(url)
    time.sleep(10)  # Wait for the page to load completely

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    articles = soup.find_all('article')
    data = []

    for article in articles:
        title_tag = article.find('a', class_="JtKRv")
        if title_tag:
            title = title_tag.get_text()
            link = article.find('a', href=True)
            if link:
                href = link['href']
                if href.startswith('/'):
                    href = 'https://news.google.com' + href
                date = datetime.now().strftime("%Y-%m-%d")
                category = classify_news(title)
                data.append({
                    'date': date,
                    'country': country,
                    'title': title,
                    'link': href,
                    'category': category
                })

    driver.quit()
    return data

# Store the results
results = []

for country, url in countries.items():
    print(f"Scraping news for {country}...")
    try:
        data = scrape_google_news(country, url)
        results.extend(data)
    except Exception as e:
        print(f"An error occurred while scraping {country}: {e}")

# Create a DataFrame and save to a CSV file
df = pd.DataFrame(results, columns=['date', 'country', 'title', 'category', 'link'])
df.to_csv('google_news_scrape_africa_oceania.csv', index=False)

print("Scraping completed. Results saved to google_news_scrape_africa_oceania.csv")
```
